```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airbnb_ny_clean.csv')
```

```python
df.head()
```

```python
# which neighbourhood group has the most AirBnbs?
# bar graph
df['neighbourhood_group'].value_counts().plot(kind='bar')
```

```python
# heat map to see correlation between different columns
corr = df.corr(method = 'kendall')
plt.figure(figsize=(15, 8))
sns.heatmap(corr, annot=True)
df.columns
```

```python
# map of where Airbnb's are mostly situated by longitude and latitude
plt.figure(figsize=(10,7))
sns.scatterplot(df.longitude, df.latitude, hue=df.availability_365)
plt.ioff()
```
